```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
    
try: 
    url="https://en.wikipedia.org/wiki/List_of_prime_ministers_of_India"
    r=requests.get(url).text
except:
    print("Please check your connection again")
soup = BeautifulSoup(r,'html.parser')
combinedtable=soup.find('table',{'class':"wikitable"})
df=pd.read_html(str(combinedtable))
df=pd.DataFrame(df[0])
df.columns=["No","No.1","Portrait","Name","Party","Took office","Left office","Timenoffice","Lok Sabha","Ministry","Appointed by","Not Needed"]
df[{"Name","Party","Took office","Left office"}]
```
