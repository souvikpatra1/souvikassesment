```
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType,StructField, StringType, IntegerType,BooleanType,DoubleType
spark = SparkSession.builder \
    .master("local[1]") \
    .appName("SparkByExamples.com") \
    .getOrCreate()
```

```

df = spark.read.option("multiline","true").json("dbfs:/FileStore/shared_uploads/souvik.patra@agilisium.com/tripdetail_json.json")
df.show()
df.printSchema()
df.toPandas().to_csv('Jsonfile.csv')
```

```
display(df)
```

```
from pyspark.sql.functions import *
df=df.withColumn("tpep_dropoff_datetime_UTC",to_utc_timestamp(df.tpep_dropoff_datetime,"PST")).withColumn("tpep_pickup_datetime_UTC",to_utc_timestamp(df.tpep_pickup_datetime,"PST"))
df.show()
```

```
df.select(df.tpep_dropoff_datetime_UTC,df.tpep_pickup_datetime_UTC).show()
```

```
display(df)
```

```
df2=df.withColumn("tpep_dropoff_datetime_IST",from_utc_timestamp(df.tpep_dropoff_datetime_UTC,"IST")).withColumn("tpep_pickup_datetime_IST",from_utc_timestamp(df.tpep_pickup_datetime_UTC,"IST"))
df2.show()
```

```
df2.select('tpep_dropoff_datetime_IST','tpep_pickup_datetime_IST').show()
df2.toPandas().to_csv('Jsonfile1.csv')
```

```
display(df2)
```

```
from pyspark.sql.functions import *
import pandas
df3 = df2.withColumn('TravelTime', expr("tpep_dropoff_datetime_IST-tpep_pickup_datetime_IST"))
df4 = df3.select('tpep_dropoff_datetime_IST','tpep_pickup_datetime_IST','TravelTime')
df4.show()
df3.show()
```
